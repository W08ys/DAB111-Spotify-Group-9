```python
import urllib.request
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Spotify_streaming_records'
```

```python
ss = requests.get(url)
ss
```

```python
page = urllib.request.urlopen(url)
soup = bs(page)
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find_all('table')
```

```python
table1 = soup.find_all('table')[0]
table1
```

```python
titles1 = table1.find_all('th')[:5]
titles1
```

```python
table1_titles = [title.text.strip() for title in titles1]
table1_titles
```

```python
df1 = pd.DataFrame(columns=table1_titles)
df1
```

```python
all_row_data = table1.find_all('tr')
all_row_data
```

```python
all_row_data
for row in all_row_data[1:101]:
    song = row.find_all('th')
    songs = [data.text.strip() for data in song]
    row_data = row.find_all('td')[:4]
    ind_row_data = [data.text.strip() for data in row_data]
    ind_row_data.insert(1 , songs.pop().strip('''"'''))
    print(ind_row_data)
    l = len(df1)
    df1.loc[l] = ind_row_data 
```

```python
df1.to_csv('SpotifyStreamingRecords_1.csv', index = False)
```
